```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('/content/smart_home_energy_consumption_large.csv')
```

```python
print("Dataset Info:")
print(df.info())
print("\nFirst 5 rows:")
print(df.head())
print("\nDescriptive Statistics:")
print(df.describe())
```

```python
print("\nMissing values per column:")
print(df.isnull().sum())
```

```python
df['DateTime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'])
```

```python
df.set_index('DateTime', inplace=True)
```

```python
df.drop(['Date', 'Time'], axis=1, inplace=True)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Appliance Type', y='Energy Consumption (kWh)',
            data=df, estimator=np.mean, errorbar=None)
plt.title('Average Energy Consumption by Appliance Type')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('energy_by_appliance.png')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Season', y='Energy Consumption (kWh)', data=df)
plt.title('Energy Consumption Distribution by Season')
plt.savefig('energy_by_season.png')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Outdoor Temperature (째C)', y='Energy Consumption (kWh)',
                hue='Appliance Type', data=df, alpha=0.6)
plt.title('Energy Consumption vs Outdoor Temperature')
plt.savefig('energy_vs_temperature.png')
plt.show()
```

```python
df['Hour'] = df.index.hour
hourly_energy = df.groupby('Hour')['Energy Consumption (kWh)'].mean()
plt.figure(figsize=(12, 6))
hourly_energy.plot(kind='line', marker='o')
plt.title('Average Hourly Energy Consumption')
plt.xlabel('Hour of Day')
plt.ylabel('Average Energy Consumption (kWh)')
plt.grid(True)
plt.savefig('hourly_energy.png')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Household Size', y='Energy Consumption (kWh)', data=df)
plt.title('Energy Consumption by Household Size')
plt.savefig('energy_by_household_size.png')
plt.show()
```

```python
appliance_energy = df.groupby('Appliance Type')['Energy Consumption (kWh)'].agg(['mean', 'std', 'count'])
appliance_energy.fillna(0, inplace=True)
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(appliance_energy)
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(scaled_data)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.savefig('elbow_method.png')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(scaled_data)
appliance_energy['Cluster'] = clusters
```

```python
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_data)
appliance_energy['PC1'] = principal_components[:, 0]
appliance_energy['PC2'] = principal_components[:, 1]

plt.figure(figsize=(10, 8))
sns.scatterplot(x='PC1', y='PC2', hue='Cluster',
                data=appliance_energy, palette='viridis', s=100)
for i, txt in enumerate(appliance_energy.index):
    plt.annotate(txt, (appliance_energy['PC1'].iloc[i], appliance_energy['PC2'].iloc[i]))
plt.title('Appliance Clusters Based on Energy Usage Patterns')
plt.savefig('appliance_clusters.png')
plt.show()
```

```python
high_energy_appliances = ['Heater', 'Air Conditioning']
for appliance in high_energy_appliances:
    appliance_data = df[df['Appliance Type'] == appliance]
    monthly_energy = appliance_data.resample('ME')['Energy Consumption (kWh)'].sum()

    plt.figure(figsize=(12, 6))
    monthly_energy.plot(title=f'Monthly Energy Consumption for {appliance}')
    plt.ylabel('Total Energy Consumption (kWh)')
    plt.savefig(f'monthly_energy_{appliance.lower().replace(" ", "_")}.png')
    plt.show()
```

```python
df_encoded = pd.get_dummies(df, columns=['Appliance Type', 'Season'])


corr_matrix = df_encoded.corr()

# Plot correlation heatmap
plt.figure(figsize=(16, 12))
sns.heatmap(corr_matrix, cmap='coolwarm', center=0, annot=True, fmt='.2f')
plt.title('Correlation Matrix of Features')
plt.tight_layout()
plt.savefig('correlation_matrix.png')
plt.show()

# Energy Consumption by Season and Appliance
plt.figure(figsize=(14, 8))
sns.barplot(x='Season', y='Energy Consumption (kWh)', hue='Appliance Type',
            data=df, estimator=np.mean, errorbar=None)
plt.title('Average Energy Consumption by Season and Appliance Type')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig('energy_by_season_appliance.png')
plt.show()


df.to_csv('processed_smart_home_energy.csv')
```
